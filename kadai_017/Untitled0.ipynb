```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def check_adult(self):
        if self.age >= 20:
          return(f"{self.name}は大人である。")
        else:
          return(f"{self.name}は大人でない。")

humanA = Human("A", 36)
humanB = Human("B", 26)
humanC = Human("C", 16)

humandataare = [humanA, humanB, humanC]
for humandatais in humandataare:
    print(humandatais.check_adult())
```
